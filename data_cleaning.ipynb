```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./datasets/House_Rent_Dataset.csv")
df
```

```python
def floor_map(value):
    if "Lower Basement" in value :
        return -2
    elif "Upper Basement" in value:
        return -1
    elif "Ground" in value:
        return 0
    else:
        return int(value.split()[0])

df["Floor"] = df["Floor"].apply(floor_map)
df
```

```python
print(set(df["Area Type"]))
print(set(df["Furnishing Status"]))
print(set(df["Tenant Preferred"]))
print(set(df["Point of Contact"]))
```

```python
df["Area Type"] = df["Area Type"].map({'Carpet Area': 0, 
                                       'Super Area': 1, 
                                       'Built Area': 2})

df["Furnishing Status"] = df["Furnishing Status"].map({'Furnished': 2, 
                                                       'Semi-Furnished': 1, 
                                                       'Unfurnished': 0})

category = pd.get_dummies(df[["Tenant Preferred", "Point of Contact", "City"]])
df = pd.concat([df, category], axis=1)
df = df.drop(columns=["Tenant Preferred", "Point of Contact", "Area Locality", "Posted On", "City"])
df
```

```python
df.to_csv("./datasets/cleaned_dataset.csv", index=False)
```
